**Csat Mosaik**

---



```python


base = pd.read_csv('/content/csat época.csv', sep=',', usecols=['data',
                                                                                       'nome_pesquisa', 
                                                                                       'tipo_pesquisa',
                                                                                       'canal', 
                                                                                        'servico', 
                                                                                        'agente',
                                                                                        'numero_contato', 
                                                                                        'nome_contato', 
                                                                                        'protocol', 
                                                                                        'Como você se sente?', 
                                                                                        'Problema Resolvido?'], encoding='utf-8').fillna('')

filtro = base['Como você se sente?']!='Nao Respondida'
base1 = base[filtro]
base1['Quantidade'] = 1
base1['Nota_Csat'] = base1['Como você se sente?']
base1['data'] = pd.to_datetime(base1['data'], dayfirst=True)

base1['Dia'] = base1['data'].dt.day
base2 = base1.copy()


# Dinamica
base_csat_novo_1 = base2.loc[:, ['Dia', 'Nota_Csat','Quantidade']].fillna('')
base_csat = pd.pivot_table(base_csat_novo_1, index=['Dia'], 
                           columns=['Nota_Csat'], 
                           values=["Quantidade"], 
                           aggfunc='sum',
                           fill_value=0)
                         
base_csat['NP'] = (base_csat['Quantidade']['4'] + base_csat['Quantidade']['5'])
base_csat['TT Pesq.'] = (base_csat['Quantidade']['1'] + base_csat['Quantidade']['2'] + base_csat['Quantidade']['3'] + base_csat['Quantidade']['4'] + base_csat['Quantidade']['5'])
base_csat['CSAT'] = round(((base_csat['NP']/ base_csat['TT Pesq.'])*100))


base_csat.head(30)

    
```

**Csat Mosaik - RCV**

---



```python
base = pd.read_csv('/content/rcv.csv', sep=',', usecols=['data',
                                                          'nome_pesquisa', 
                                                          'tipo_pesquisa',
                                                          'canal', 
                                                          'servico', 
                                                          'agente',
                                                          'numero_contato', 
                                                          'nome_contato', 
                                                          'protocol', 
                                                          'Como você se sente?', 
                                                          'Problema Resolvido?'], encoding='utf-8').fillna('')

filtro = base['Como você se sente?']!='Nao Respondida'
base1 = base[filtro]
base1['Quantidade'] = 1
base1['Nota_Csat'] = base1['Como você se sente?']
base1['data'] = pd.to_datetime(base1['data'], dayfirst=True)

base1['Dia'] = base1['data'].dt.day
base2 = base1.copy()


def val_servico(base2):
  if base2['servico'] == 'Canal Vermelho':
      return "Resolve"
  elif base2['servico'] == 'Resolve com você':
    return "Resolve"
  else:
      return "Resolve Já"
base2['Agrupamento_servico'] = base2.apply(val_servico, axis=1)


# Dinamica
base_csat_novo_1 = base2.loc[:, ['Agrupamento_servico','Dia', 'Nota_Csat','Quantidade']].fillna('')
base_csat = pd.pivot_table(base_csat_novo_1, index=['Agrupamento_servico', 'Dia'], 
                           columns=['Nota_Csat'], 
                           values=["Quantidade"], 
                           aggfunc='sum',
                           fill_value=0)
                         
base_csat['NP'] = (base_csat['Quantidade']['4'] + base_csat['Quantidade']['5'])
base_csat['TT Pesq.'] = (base_csat['Quantidade']['1'] + base_csat['Quantidade']['2'] + base_csat['Quantidade']['3'] + base_csat['Quantidade']['4'] + base_csat['Quantidade']['5'])
base_csat['CSAT'] = round(((base_csat['NP']/ base_csat['TT Pesq.'])*100))

base_csat.head()
```

**Csat Solucx Magalu**

---



```python
base = pd.read_csv('/content/csat_solucx_email.csv', sep=',')
base['criado_em'] = pd.to_datetime(base['criado_em'], dayfirst=True)
base['criado_em'].dt.strftime('%Y-%m-%d')
base['Quantidade'] = 1

base_csat = pd.pivot_table(base, index=['criado_em'], 
                           columns=['nota'], 
                           values=["Quantidade"], 
                           aggfunc='count',
                           fill_value=0)


base_csat['NP'] = (base_csat['Quantidade'][4.0] + base_csat['Quantidade'][5.0])
base_csat['TT pesquisas'] = (base_csat['Quantidade'][1.0] + 
                             base_csat['Quantidade'][2.0] + 
                             base_csat['Quantidade'][3.0] + 
                             base_csat['Quantidade'][4.0] + 
                             base_csat['Quantidade'][5.0])

base_csat['CSAT'] = round(((base_csat['NP']/ base_csat['TT pesquisas'])*100))


base_csat.head()
```
