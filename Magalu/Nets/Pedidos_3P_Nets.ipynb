```python
base_nse = pd.read_excel(r'C:\Users\mo_duarte\Desktop\Marketplace\pedidos\Base_pedidos_nets\agosto\SELLER NSE.xlsx', skiprows=2).fillna('')
base_nse.rename(columns={'Seller': 'LOJISTA'}, inplace=True)
base_nse.head()
```

```python
base_tt = pd.merge(base_nova_1, base_nse, how='left', on='LOJISTA')

base_tratada = base_tt.drop_duplicates(subset='COD_PEDIDO', keep='first').fillna('')

base_tratada_v1 = base_tratada.query('ID==""')

base_tratada_v1['DATA_ENTREGA'] = pd.to_datetime(base_tratada_v1['DATA_ENTREGA'], format='%Y%m%d', infer_datetime_format=True)
base_tratada_v1['DATA_DESPACHO'] = pd.to_datetime(base_tratada_v1['DATA_ENTREGA'], format='%Y%m%d', infer_datetime_format=True)

def status_id(base_tratada_v1):
    if base_tratada_v1['STATUS'] == 'DESPACHADO':
        return "S"
    elif base_tratada_v1['STATUS'] == 'ENTREGUE':
        return "S"
    elif base_tratada_v1['STATUS'] == 'FATURADO':
        return "S"
    elif base_tratada_v1['STATUS'] == 'PRONTO PARA FATURAR':
        return "S"
    else:
        return ""

base_tratada_v1['Val_Status'] = base_tratada_v1.apply(status_id, axis=1)
       
base_tratada_v2 = base_tratada_v1.query('Val_Status=="S"')  

base_tratada_v3 = base_tratada_v2[['LOJISTA', 'COD_PEDIDO_LOJA','COD_PEDIDO', 'DATA_DESPACHO','DATA_ENTREGA',  'Dt_Pedidos', 'Dt_Aprov', 'DIAS_PRAZO', 'STATUS']]

          
base_tratada_v3.tail()


```
